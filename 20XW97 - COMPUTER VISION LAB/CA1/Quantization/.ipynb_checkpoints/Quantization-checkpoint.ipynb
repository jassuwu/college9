# Quantization of the signal

## Objective: Implementation of a scalar quantizer



Quantization is part of the following flow:

Analog Signal -> Sampling -> **Quantization** -> DS


- **Quantization**: mapping a large set of values to a smaller set of values.
- **Advantage**: compression
- **Disadvantage**: loss of uniqueness/information. error is inevitable. non-linear mapping.
    - due to a many to one mapping. unique when there's a one to one mapping
    
    
E.g. of Quantization: Grading system for grades -> `{"60 to 70"-> C, "70 to 80"-> B, "80 to 90"-> A}`


### Quantization terminology and steps
1. **DR**: Dynamic Range i.e. difference between max and min value in the signal
2. **Bit Budget**: `b`
3. **L**: 2<sup>b</sup> (Reconstruction level)
4. **S**: DR/L (Quantization step)

Quantization gives us an appx. of the signal.

**Now,**

**i/p**: sine wave `x` <br>
**o/p**: quantized sine wave `y = Q(x)`


**Solution,**

1. Generate sine wave
2. Design the Quantizer (Uniform Quantizer, here.)
3. Perform the Quantization
4. Visualize the result

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Generation of a sine wave
t = np.linspace(0, 1, 100)
x = np.sin(2*np.pi*5*t)
plt.plot(t, x)
```

```python
# Step 2 and the crucial: Designing the (Uniform) Quantizer
dr = np.max(x) - np.min(x)
b = 1
l = 2 ** b
q = dr / l
```

```python
# Step 3: Perform Quantization
y = np.floor(x/q)*q
```

```python
# Step 4: Visualize the results
plt.plot(t, x)
plt.plot(t, y)
plt.legend(['i/p', 'o/p'], loc=1)
```

That output looks nothing like the input. That's because the bit budget is very low.

Let's try increasing the bit budget slowly.

```python
dr = np.max(x) - np.min(x)
for b in range(1, 9):
    l = 2 ** b
    q = dr / l
    y = np.floor(x/q)*q
    plt.plot(t, x)
    plt.plot(t, y)
    plt.show()
```

First thing to ask about when talking about quantization: `bit budget`

## Uniform Quantization of an Image

```python
import cv2
```

```python
# 1 - Changes
x = cv2.imread("cameraman.tif", 0)
# 2 - Nothing changes
dr = np.max(x) - np.min(x)
b = 1
l = 2 ** b
q = dr / l
# 3 - Nothing changes
y = np.floor(x/q)*q
# 4 - Nothing really changes, we're still only visualizing the result
plt.subplot(1, 2, 1), plt.imshow(x, cmap='gray')
plt.subplot(1, 2, 2), plt.imshow(y, cmap='gray')
```

So without changing anything about the quantizer, different types of inputs can be used for the same.

### MASK PROCESSING / 2D Convolution in between the technical difficulties

i/p -> **Kernel/Mask** -> o/p

**Objective** : Learn the impact of kernel on input image

**x(m,n) * (m,n) = x(m,n)**

#### Steps

1. Read the input image
2. Create the mask
3. Perform 2D Convolution
4. Display the result

```python
x = cv2.imread('cameraman.tif')
h = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]]) # Identity function
y = cv2.filter2D(x, -1, h)
plt.subplot(1, 2, 1), plt.imshow(x, cmap='gray')
plt.subplot(1, 2, 2), plt.imshow(y, cmap='gray')
```
