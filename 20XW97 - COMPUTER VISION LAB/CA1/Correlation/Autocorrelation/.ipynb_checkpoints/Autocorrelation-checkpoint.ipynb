### 1. Autocorrelation -> Template Matching


```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import signal
import cv2 as cv
```

```python
x = np.array([[1, 2], [3, 4]])
z = signal.correlate2d(x, x, mode="full")
z
```

### Cross Correlation

```python
# Creating templates and images
template = np.zeros((8, 8))
template[4, :] = 1

img1 = np.zeros((8, 8))
img1[4, 4] = 1

img2 = np.zeros((8, 8))
img2[:, 4] = 1

# img3 = np.zeros((8, 8))
# x, y = np.meshgrid(np.arange(8), np.arange(8))
# img3[x == y] = 1
img3 = np.eye(8)

img4 = template

plt.figure(figsize=(8, 4))

plt.subplot(151), plt.imshow(template, cmap="gray"), plt.title("template"), plt.axis("off")
plt.subplot(152), plt.imshow(img1, cmap="gray"), plt.title("img1"), plt.axis("off")
plt.subplot(153), plt.imshow(img2, cmap="gray"), plt.title("img2"), plt.axis("off")
plt.subplot(154), plt.imshow(img3, cmap="gray"), plt.title("img3"), plt.axis("off")
plt.subplot(155), plt.imshow(img4, cmap="gray"), plt.title("img4"), plt.axis("off")

plt.tight_layout()
```

```python
# Perform Cross correlation
cc1 = signal.correlate2d(template, img1, mode="full")
cc2 = signal.correlate2d(template, img2, mode="full")
cc3 = signal.correlate2d(template, img3, mode="full")
cc4 = signal.correlate2d(template, img4, mode="full")

cc = [cc1, cc2, cc3, cc4]
m = list(map(np.max, cc))

best_match = np.argmax(m)
best_match
```

We will always get a **"best"** match. But we also need a **"no match"** result.

### 2. Visual Perception -> Optical Illusion -> Simultaneous Contrast

- Change in gray level

Simultaneous Contrast 👇

```python
img1 = np.zeros((256, 256))
img1[108:138, 108:138] = 128

img2 = np.zeros((256, 256))
img2 = img2 + 80
img2[108:138, 108:138] = 128

img3 = np.zeros((256, 256))
img3 = img2 + 200
img3[108:138, 108:138] = 128

plt.figure(figsize=(8, 4))

plt.subplot(131), plt.imshow(img1, cmap="gray", vmin=0, vmax=255), plt.title("img1"), plt.axis("off")
plt.subplot(132), plt.imshow(img2, cmap="gray", vmin=0, vmax=255), plt.title("img2"), plt.axis("off")
plt.subplot(133), plt.imshow(img3, cmap="gray", vmin=0, vmax=255), plt.title("img3"), plt.axis("off")

plt.tight_layout()
```

### Kofka Benuzzi ring - A good example for Simultaneous Contrast

```python
# 1. Different background
img = np.zeros((256, 256))
img[:, :128] = 255

# 2. Ring Structure
m, n = img.shape
r = 75

for i in range(m):
    for j in range(n):
        d = ((i-128)**2 + (j-128)**2)/(r**2)
        if d > 0.5 and d < 1:
            img[i, j] = 128
            
img[:, 118:138] = 255
            
plt.imshow(img, cmap="gray", vmin=0, vmax=255), plt.title("Kofta Benuzzi Ring"), plt.axis("off")
```

### Gray level transformation

In a system, <br>
x-axis => i/p gray level <br>
y-axis => o/p gray level <br>

f(m, n) -> `Transformation System` -> g(m, n)?

Some transformations:
- Identity `(0 -> 255) => (0 -> 255)`
- GL Inversion `(0 -> 255) => (255 -> 0)`

```python
img = cv.imread("cameraman.tif", cv.IMREAD_GRAYSCALE)
plt.figure(figsize=(8, 4))

plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title("img"), plt.axis("off")
plt.subplot(122), plt.imshow(255 - img, cmap="gray"), plt.title("GL inverted img"), plt.axis("off")

plt.tight_layout()
```
