```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

### Task 1
Generate piecewise linear contrast stretching without using `np.vectorize`.


https://www.geeksforgeeks.org/python-intensity-transformation-operations-on-images/

```python
def piecewise_linear_contrast_stretching(img, r1, s1, r2, s2):
    """
    Apply piecewise linear contrast stretching to an image.

    Parameters:
    img (numpy.ndarray): Input grayscale image.
    r1 (int): Intensity value for the first input threshold.
    s1 (int): Intensity value for the first output threshold.
    r2 (int): Intensity value for the second input threshold.
    s2 (int): Intensity value for the second output threshold.

    Returns:
    numpy.ndarray: Contrast stretched image.
    """
    # Create an output image
    stretched_img = np.zeros_like(img)

    # Apply piecewise linear transformation
    # Segment 1: (0 <= pix <= r1)
    mask1 = (img <= r1)
    stretched_img[mask1] = (s1 / r1) * img[mask1]

    # Segment 2: (r1 < pix <= r2)
    mask2 = (img > r1) & (img <= r2)
    stretched_img[mask2] = ((s2 - s1) / (r2 - r1)) * (img[mask2] - r1) + s1

    # Segment 3: (r2 < pix <= 255)
    mask3 = (img > r2)
    stretched_img[mask3] = ((255 - s2) / (255 - r2)) * (img[mask3] - r2) + s2

    return stretched_img

img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)

r1 = 70
s1 = 0
r2 = 140
s2 = 255

stretched_img = piecewise_linear_contrast_stretching(img, r1, s1, r2, s2)

plt.figure(figsize=(12, 6))

plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(132), plt.imshow(piecewise_linear_contrast_stretching(img, 1, 0, 254, 255), cmap='gray'), plt.title('Contrast stretched ((0, 0) and (255, 255))'), plt.axis('off')
plt.subplot(133), plt.imshow(piecewise_linear_contrast_stretching(img, 70, 0, 140, 255), cmap='gray'), plt.title('Contrast stretched ((70, 0) and (140, 255))'), plt.axis('off')

plt.tight_layout()
```

### Task 2
Plotting of a histogram of an image using `matplotlib` and `cv2`.

```python
img = cv.imread("cameraman.tif", cv.IMREAD_GRAYSCALE)
hist = cv.calcHist([img], [cv.IMREAD_GRAYSCALE], None, [255], (0, 256))

plt.figure(figsize=(12, 6))

plt.subplot(121), plt.hist(img.flatten()), plt.title("matplotlib")
plt.subplot(122), plt.plot(hist), plt.title("opencv")

plt.tight_layout()
```

### Task 3
Read a cameraman image of 256 x 256. Shuffle all the pixels in the image and generate a shuffle image.
Plot histogram for original image.
Plot histogram for shuffle image.
Observe whether both the images are same.

```python
img = cv.imread("cameraman.tif", cv.IMREAD_GRAYSCALE)

h, w = img.shape
permutation = np.random.permutation(h * w)
shuffled_img = img.flatten()[permutation].reshape(h, w)

plt.figure(figsize=(12, 6))

plt.subplot(121), plt.hist(img.flatten()), plt.title("img")
plt.subplot(122), plt.hist(shuffled_img.flatten()), plt.title("shuffled_img")

plt.tight_layout()
```

**Inference:** They have the same histograms.

### Task 4
Using python prove that histogram equivalization is an Idempotent operator.

```python
img = cv.imread('cameraman.tif', cv.IMREAD_GRAYSCALE)

def apply_equalization_n_times(img, n):
    result = img
    for _ in range(n):
        result = cv.equalizeHist(result)
    return result

n_times = np.random.randint(100, size=1)[0]

equalized_first = apply_equalization_n_times(img, 1)
equalized_n_times = apply_equalization_n_times(img, n_times) # Apply it a random amount of times

plt.figure(figsize=(20, 4))

plt.subplot(131), plt.hist(img.flatten(), bins=256, range=(0, 256)), plt.title('Original Image Histogram')
plt.subplot(132), plt.hist(equalized_first.flatten(), bins=256, range=(0, 256)), plt.title('Equalized Once Histogram')
plt.subplot(133), plt.hist(equalized_n_times.flatten(), bins=256, range=(0, 256)), plt.title(f'Equalized {n_times} Times Histogram')

plt.show()
```
