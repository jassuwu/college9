```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

# Task 1

```python
def nullify_ft_coefficients(img, percentage):
    IMG = np.fft.fft2(img)
    
    total_coefficients = IMG.size
    num_to_nullify = int(total_coefficients * (percentage / 100.0))
    
    indices = np.random.choice(total_coefficients, num_to_nullify, replace=False)
    IMG_flat = IMG.flatten()
    IMG_flat[indices] = 0
    IMG = IMG_flat.reshape(IMG.shape)
    
    return IMG

img = cv.imread("cameraman.tif", cv.IMREAD_GRAYSCALE)
percentage = float(input("Enter the percentage of Fourier coefficients to nullify: "))
modified_IMG = nullify_ft_coefficients(img, percentage)
rimg = np.fft.ifft2(modified_IMG)
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title("Original image")
plt.subplot(122), plt.imshow(np.abs(rimg), cmap="gray"), plt.title(f"Reconstructed image ({percentage}% nullified)")
plt.tight_layout()
plt.show()
```

# Task 2

```python
import numpy as np
import matplotlib.pyplot as plt

N = 4
dft_matrix = np.fft.fft(np.eye(N))

eval, evec = np.linalg.eig(dft_matrix)

def check_orthogonality(eigenvectors):
    for i in range(len(eigenvectors)):
        for j in range(i + 1, len(eigenvectors)):
            dot_product = np.dot(eigenvectors[:, i].conj().T, eigenvectors[:, j])
            print(f"Dot product of eigenvector {i+1} and eigenvector {j+1}: {dot_product:.5f}")

check_orthogonality(evec)

fig, axs = plt.subplots(2, 3, figsize=(10, 6))

colors = ['r', 'b']
pairs = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]
for idx, (i, j) in enumerate(pairs):
    row, col = divmod(idx, 3)
    ax = axs[row, col]
    
    for k in range(2):  # Compare first two eigenvectors
        v = evec[:, k].real  # Take real part of the eigenvector
        ax.quiver(0, 0, v[i], v[j], color=colors[k], angles='xy', scale_units='xy', scale=1, label=f"Vec {k+1}")
    
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_title(f"Components {i+1} vs {j+1}")
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(True)

plt.tight_layout()
plt.show()
```
