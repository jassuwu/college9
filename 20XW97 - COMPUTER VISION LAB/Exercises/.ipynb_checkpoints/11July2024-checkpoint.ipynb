```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sounddevice as sd
import cv2
import os
import scipy.io as spio
```

## Exercise 1
- Read an image
- Store them in different formats
- Bar chart their storage sizes.

```python
# Step 1: Read an Image
image = cv2.imread("cameraman.tif")

# Step 2: Store the Image in Different Formats
formats = ['jpeg', 'png', 'bmp', 'webp']
sizes = []

# Create a directory to save the images
os.makedirs('11july24/converted_images', exist_ok=True)

for fmt in formats:
    save_path = f'11july24/converted_images/cameraman.{fmt}'
    if fmt == 'jpeg':
        cv2.imwrite(save_path, image, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
    elif fmt == 'png':
        cv2.imwrite(save_path, image, [int(cv2.IMWRITE_PNG_COMPRESSION), 3])
    elif fmt == 'bmp':
        cv2.imwrite(save_path, image)
    elif fmt == 'webp':
        cv2.imwrite(save_path, image)
    sizes.append(os.path.getsize(save_path))

# Step 3: Bar Chart Their Storage Sizes
plt.bar(formats, sizes, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Image Format')
plt.ylabel('File Size (bytes)')
plt.title('Storage Sizes of Different Image Formats')
plt.show()
```

## Exercise 3
- Read an image
- Scramble/Shuffle the pixels
- Try to build back the original

```python
# Step 1: Read the Image
image_path = 'cameraman.tif'  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Step 2: Scramble/Shuffle the Pixels
# Get the shape of the image
h, w = image.shape

# Flatten the image to a 1D array
flattened_image = image.flatten()

# Generate a random permutation of indices
permutation = np.random.permutation(h * w)

# Scramble the pixels using the permutation
scrambled_image = flattened_image[permutation].reshape(h, w)

# Step 3: Attempt to Rebuild the Original Image
# We'll sort the scrambled image using the permutation to try and rebuild the original
rebuild_indices = np.argsort(permutation)
reconstructed_image = scrambled_image.flatten()[rebuild_indices].reshape(h, w)

# Plot the original, scrambled, and reconstructed images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Scrambled Image")
plt.imshow(scrambled_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Reconstructed Image")
plt.imshow(reconstructed_image, cmap='gray')

plt.show()
```

## Exercise 4
- Get voice, store in .wav
- Display the voice
- fs=8000, duration=10seconds

```python
# Parameters
fs = 8000  # Sampling frequency
duration = 10  # Duration in seconds
filename = "voice.wav"
```

```python
# Step 1: Record the voice
print("Recording...")
recorded_data = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='int16')
sd.wait()  # Wait until the recording is finished
print("Recording finished")
```

```python
# Step 2: Save the recording as a .wav file
spio.wavfile.write(filename, fs, recorded_data)
```

```python
# Step 3: Display the voice signal
fs, data = spio.wavfile.read(filename)
plt.plot(data)
plt.title("Recorded Voice Signal")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Step 4: Play back the recorded audio
sd.play(data, fs)
sd.wait()
```
