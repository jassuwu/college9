# Drawing the Indian Flag because yeah.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H, W = 60, 90
```

```python
flag = np.ones((H, W, 3), dtype=np.uint8) * 255
```

```python
flag[:20, :] = [255, 103, 31]
# flag[20:40, :] is already white
flag[40:60, :] = [4, 106, 56]
```

```python
center = (W // 2 - 0.5, H // 2 - 0.5)
radius = 10
```

```python
fig = plt.figure(figsize=(60, 90))
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * np.sin(theta)
```

```python
for i in range(len(x)-1):
    plt.plot([x[i], x[i+1]], [y[i], y[i+1]], color='#06038D')
    
for i in range(24):
    angle = i * (2 * np.pi / 24)
    x1 = center[0] + radius * np.cos(angle)
    y1 = center[1] + radius * np.sin(angle)
    plt.plot([center[0], x1], [center[1], y1], color='#06038D')

plt.imshow(flag)
plt.axis('off')
plt.show()
```
