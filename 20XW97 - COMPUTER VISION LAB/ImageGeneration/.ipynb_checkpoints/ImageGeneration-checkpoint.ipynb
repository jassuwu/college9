### CIRCLE IN A SQUARE
- Importance of circle: minimum point of contact.
- We go back to the equation of the circle. `(x-a)^2 + (y-b)^2 = r^2`

```python
img = np.zeros((256, 256))
r=20

for x in range(256):
    for y in range(256):
        if (x-128)**2 + (y-128)**2 <= r**2:
            img[x,y]=255

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

### CONCENTRIC CIRCLES EXERCISE

```python
img = np.zeros((256, 256))

for r in [10, 20, 30]:
    for x in range(256):
        for y in range(256):
            if (x-128)**2 + (y-128)**2 == r**2:
                img[x,y]=255

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

### ELLIPSE
- The formula: `((x-h)^2/a^2) + ((y-k)^2/b^2) = 1`

```python
img = np.zeros((256, 256))

r = 10
a = 10
b = 5

for x in range(256):
    for y in range(256):
        if ((x-128)/b)**2 + ((y-128)/a)**2 <= r**2:
            img[x,y]=255

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

### GRADIENT IMAGE
- varies slowly from a value to another value

```python
img = np.zeros((256, 256))

img[:,:] = np.arange(0, 256, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.plot(img[128,:])
plt.show()
```

Inference: We can see the gradual change, so `y=mx`

How to generate something like an Zoneplate image?

```python
x = y = np.linspace(-128, 128, 256)
X, Y = np.meshgrid(x, y)
Z = np.cos(256*np.pi*(X**2+Y**2))

plt.imshow(Z, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.plot(Z[128,:])
plt.show()
```

Explore the above code.

### Zoneplate image
- Pattern is repetitive. It's periodic. This is the first characteristic that one should notice.
- It is symmetric.
- cosine fn. shows `even symmetry`. zoneplate does.





```python
x = y = np.linspace(-128, 128, 256)
X, Y = np.meshgrid(x, y)
Z = np.sin(256*np.pi*(X**2+Y**2))

plt.imshow(Z, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.plot(Z[128,:])
plt.show()
```

### BLURRED CIRCLE

```python
x = y = np.linspace(-128, 128, 256)
X, Y = np.meshgrid(x, y)
D = np.sqrt(X**2+Y**2)
D0 = 20

img=np.exp(-D**2/(2*D0**2))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.plot(img[128,:])
plt.show()
```
