```python
# 3
x, y, z = sp.symbols('x y z')
sp.init_printing(use_unicode=True)
```

```python
# 3a
y = x ** 2
ydiff2 = sp.diff(y, x, 2)

s = sp.plotting.plot(y, show=False)
s.extend(sp.plotting.plot(ydiff2, show=False))
print(f"Second ord. derivative: {ydiff2}\nConvex Range: {sp.solveset(ydiff2>=0, x, sp.Reals)}\nConcave Range: {sp.solveset(ydiff2<=0, x, sp.Reals)}")
s.show()
```

```python
# 4

unit_matrix = np.array([[1, 1], [1, 1]])
square_not_symmetric = np.array([[1, 2], [3, 4]])
symmetric = np.array([[1, 2], [2, 3]])
rectangular = np.array([[1, 2, 3, 4]])
identity = np.array([[1, 0], [0, 1]])

all_matrices = [unit_matrix, square_not_symmetric, symmetric, rectangular, identity]

# a) Determinant
print("Determinant")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.det(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# b) Inverse Matrix
print("Inverse Matrix")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.inv(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# c) Matrix rank
print("Matrix rank")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.matrix_rank(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# d) Matrix Orthogonality
print("Matrix Orthogonality")
for m in all_matrices:
    try:
        print(m, ": ", np.allclose(np.linalg.inv(m), m.T))
    except Exception as e:
        print("Error evaluating",m, ":", e)

print("=======================================")

# e) Covariance
print("Covariance")
for m in all_matrices:
    try:
        print(m, ": ", np.cov(m))
    except Exception as e:
        print("Error evaluating",m, ":", e)

print("=======================================")

# e) Correlation
print("Correlation")
for m in all_matrices:
    try:
        print(m, ": ", np.corrcoef(m))
    except Exception as e:
        print("Error evaluating",m, ":", e)
```

```python
# 5 IRIS DATASET
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

X, y
```

```python
# a
print("Dataset rank:", np.linalg.matrix_rank(X))
# Since the rank = no. of features, the features are linearly independent.

print("nullity: ", X.shape[1] - np.linalg.matrix_rank(X))
```

```python
# b

print("Mean: ", X.mean())
print("=======================================")
print("Standard Deviation:", X.std())
print("=======================================")
print("Covariance:", X.cov())
print("=======================================")
print("Correlation:", X.corr())
print("=======================================")
print("Skewness:", X.skew())
print("=======================================")
print("Kurtosis:", X.kurt())
```

```python
# c
print("Mean adjusted values:")
X.mean() - X
```

```python
# f

fig, ax = plt.subplots(4, 3, figsize=(12, 12))
for idx, f1 in enumerate(X.columns):
    count = 0
    for f2 in X.columns:
        if f1 != f2:
            ax[idx, count].scatter(X[f1], X[f2], c=y, cmap='viridis', alpha=0.5)
            ax[idx, count].set_title(f'{f1} vs {f2}')
            ax[idx, count].set_xlabel(f1)
            ax[idx, count].set_ylabel(f2)
            count += 1

plt.show()
```
