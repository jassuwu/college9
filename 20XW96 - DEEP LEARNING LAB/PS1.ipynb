```python
# 4

unit_matrix = np.array([[1, 1], [1, 1]])
square_not_symmetric = np.array([[1, 2], [3, 4]])
symmetric = np.array([[1, 2], [2, 3]])
rectangular = np.array([[1, 2, 3, 4]])
identity = np.array([[1, 0], [0, 1]])

all_matrices = [unit_matrix, square_not_symmetric, symmetric, rectangular, identity]

# a) Determinant
print("Determinant")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.det(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# b) Inverse Matrix
print("Inverse Matrix")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.inv(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# c) Matrix rank
print("Matrix rank")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.matrix_rank(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# d) Matrix Orthogonality
print("Matrix Orthogonality")
for m in all_matrices:
    try:
        print(m, ": ", np.allclose(np.linalg.inv(m), m.T))
    except Exception as e:
        print("Error evaluating",m, ":", e)

print("=======================================")

# e) Covariance
print("Covariance")
for m in all_matrices:
    try:
        print(m, ": ", np.cov(m))
    except Exception as e:
        print("Error evaluating",m, ":", e)

print("=======================================")

# e) Correlation
print("Correlation")
for m in all_matrices:
    try:
        print(m, ": ", np.corrcoef(m))
    except Exception as e:
        print("Error evaluating",m, ":", e)
```

```python
# k. Calculate the eigen values and eigen vectors

corr_eval, corr_evec = np.linalg.eig(X.corr())
cov_eval, cov_evec = np.linalg.eig(X.cov())

(corr_eval, cov_eval, corr_evec, cov_evec)
```

```python
# l. Find the L2 Norm of the above Eigen vectors

corr_evec_l2_norm = np.linalg.norm(corr_evec, 2)
cov_evec_l2_norm = np.linalg.norm(cov_evec, 2)

(corr_evec_l2_norm, cov_evec_l2_norm)
```
