```python
# 3
x, y, z = sp.symbols('x y z')
sp.init_printing(use_unicode=True)
```

```python
# 3a
y = x ** 2
ydiff2 = sp.diff(y, x, 2)

s = sp.plotting.plot(y, show=False)
s.extend(sp.plotting.plot(ydiff2, show=False))
print(f"Second ord. derivative: {ydiff2}\nConvex Range: {sp.solveset(ydiff2>=0, x, sp.Reals)}\nConcave Range: {sp.solveset(ydiff2<=0, x, sp.Reals)}")
s.show()
```

```python
# 4

unit_matrix = np.array([[1, 1], [1, 1]])
square_not_symmetric = np.array([[1, 2], [3, 4]])
symmetric = np.array([[1, 2], [2, 3]])
rectangular = np.array([[1, 2, 3, 4]])
identity = np.array([[1, 0], [0, 1]])

all_matrices = [unit_matrix, square_not_symmetric, symmetric, rectangular, identity]

# a) Determinant
print("Determinant")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.det(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# b) Inverse Matrix
print("Inverse Matrix")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.inv(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# c) Matrix rank
print("Matrix rank")
for m in all_matrices:
    try:
        print(m, ": ", np.linalg.matrix_rank(m))
    except Exception as e:
        print("Error evaluating", m, ":", e)

print("=======================================")

# d) Matrix Orthogonality
print("Matrix Orthogonality")
for m in all_matrices:
    try:
        print(m, ": ", np.allclose(np.linalg.inv(m), m.T))
    except Exception as e:
        print("Error evaluating",m, ":", e)

print("=======================================")

# e) Covariance
print("Covariance")
for m in all_matrices:
    try:
        print(m, ": ", np.cov(m))
    except Exception as e:
        print("Error evaluating",m, ":", e)

print("=======================================")

# e) Correlation
print("Correlation")
for m in all_matrices:
    try:
        print(m, ": ", np.corrcoef(m))
    except Exception as e:
        print("Error evaluating",m, ":", e)
```
